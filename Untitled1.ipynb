Los datos utilizados en España en llamas, tanto en el mapa de incendios como en muchos de los artículos, proceden de la Estadística General de Incendios Forestales (EGIF), que elabora en el Centro de Coordinación de la Información Nacional de Incendios Forestales (CCINIF) a partir de la información anual suministrada por las comunidades autónomas. Y que desde Civio llevamos años reclamando en cada ejercicio. Esta procede directamente de los partes de incendio forestal, que se rellenan in situ para cada cada fuego, y que contienen más de 150 campos de información. Este parte es consensuado por todas las administraciones responsables en el Grupo de Trabajo de Estadísticas del Comité de Lucha contra Incendios Forestales (CLIF). Cuando los datos de las comunidades llegan al CCINIF, este los procesa y consolida, dando forma a la EGIF. Una vez que se completa la información de cada año, es trasladada a la Comisión Europea para nutrir la base de datos descentralizada de la Unión. Este proceso de elaboración hace que los datos disponibles más recientes sean, habitualmente, de dos años anteriores al año en curso.

Para tratar estos datos, hemos convertido la base de datos MS Access del ministerio a MySQL utilizando la utilidad de conversión desarrollada por Cynergi. Posteriormente, convertimos las coordenadas de localización de los incendios y de los municipios de UTM a coordenadas geográficas WGS 84 usando la herramienta desarrollada por José Guerrero. Por último, se convirtieron los tiempos de control y extinción de fecha y hora a unidades de minutos.

En el análisis de la información contenida en la base de datos se detectaron las siguientes carencias para realizar un tratamiento homogéneo de los datos:

La causalidad de los incendios forestales no es cierta, sino supuesta en más del 55% de los casos.

La geolocalización de los incendios no está especificada en casi un 18% de los incendios.

La información sobre pérdidas económicas y gastos es muy escasa, lo cual hace muy difícil analizar el coste real de los incendios que han ocurrido en España. Más de un 30% de los partes de incendios no proporcionan datos de gastos de extinción y casi un 9% no dispone de datos de pérdidas económicas.

Además, los datos que ofrece el Ministerio pueden contener errores. Desde Civio no podemos corroborar que todos y cada uno de esos datos, incendio a incendio, son exactos, por lo que tenemos que trabajar sobre la información oficial disponible.

El fichero CSV descargado contiene los campos descritos a continuación. Algunos campos (provincia, causa…) contienen códigos, definidos en estas tablas adicionales. El sistema de coordenadas es WGS84 y los importes son en euros.


* id	Identificador del incendio
* superficie	Superficie forestal quemada en hectáreas
* fecha	Fecha de detección del incendio (formato yyyy-mm-dd)
* lat	Latitud geográfica del origen del incendio
* lng	Longitud geográfica del origen del incendio
* latlng_explicit	Indica si se dispone de coordenadas geográficas del incendio (1) o se han usado las coordenadas del municipio de origen del incendio (0)
* comunidad	Identificador de la CC.AA.
* provincia	Identificador de la provincia
* municipio	Nombre del municipio
* causa	Identificador de la causa del incendio
* causa_supuesta	Es ‘1’ si la causa es supuesta, si no en blanco
* causa_desc	Identificador de la descripción de la causa del incendio
* muertos	Número de muertos en el incendio
* heridos	Número de heridos en el incendio
* time_ctrl	Tiempo transcurrido hasta entrar en fase de control del incendio (en minutos)
* time_ext	Tiempo transcurrido hasta la extinción del incendio (en minutos)
* personal	Número de personas que han participado en la extinción del incendio (incluye técnicos, agentes forestales, brigadas, bomberos, voluntarios, guardias civiles y ejército)
* medios	Número de medios terrestres y aéreos que han participado en la extinción del incendio (incluye autobombas, bulldozers, tractores, aviones y otros)
* gastos	Gastos de extinción asociados al incendio tal y como figuran en EGIF
* perdidas	Pérdidas económicas asociadas al incendio tal y como figuran en EGIF

```python
#remove NaN rows in lat and lng because they are only 24 rows 

df.drop(df[(df["lng"].isna())].index, inplace = True)
#df = df["lat"].dropna()
```

```python
df.isna().sum()
```

```python
#drop id; adds no info and we have another index now which is the date
df = df.drop(["id"], axis = 1)

#drop latlng_explicit; only says if the coordinates were taken from the town or not
df = df.drop(["latlng_explicit"], axis = 1)

#drop idmunicipio; we already have the name of the municipality
df = df.drop(["idmunicipio"], axis = 1)

#drop gastos; although this variable would be very insightful, it has too many NaN values that cannot be eliminated or filled
df = df.drop(["gastos"], axis = 1)

#drop perdidas; same as above
df = df.drop(["perdidas"], axis = 1)
```

```python
#in causa_supuesta, it's either 1 or NaN; so we fill the NaN we 0
df["causa_supuesta"] = df.causa_supuesta.fillna(0)
```

```python
df.head()
```

```python
causa_map = {1 : "rayo", 2 : "accidente/negligencia", 3 : "accidente/negligencia", 4 : "intencionado", 5 : "desconocido", 
            6 : "reproducido"}
df["causa2"] = df["causa"].map(causa_map).astype('str')
```

```python
#rename values to strings to make dummies and because 2 and 3 stand for the same

df["causa"] = df["causa"].replace([1, 4, 5, 6],["rayo", "intencionado", "desconocido", "reproducido"])

df["causa"] = df["causa"].replace([2, 3],"accidente/negligencia")
```

```python
#create variable geohash
df["geohash"] = df.apply(lambda x: pgh.encode(x.lat, x.lng), axis = 1)
```

```python
#create new column with types of wildfires
# conato: less than 1 ha
# gran incendio forestal (GIF): 500 ha or more
# incendio: etween 1 and 499 ha

def new_col_tipo(row):
    if row["superficie"] < 1:
        return "conato"
    elif row["superficie"] >= 500:
        return "GIF"
    else:
        return "incendio"


df["tipo"] = df.apply(new_col_tipo, axis =1)
```
